```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("/Users/bharathi/Documents/sundar/olympics.csv")
data.columns = data.columns.str.strip()

def plot_medals(data, region):

 data_EU = data[data["region"] == region]
 fig, ax = plt.subplots(figsize=(10, 6))
 ax.scatter(data_EU["gold"], data_EU["total medals"], color='red', marker='o', s=30, alpha=0.9, label="Gold") 
 ax.scatter(data_EU["silver"], data_EU["total medals"], color='blue', marker='o', s=30, alpha=0.6, label="Silver")
 ax.scatter(data_EU["bronze"], data_EU["total medals"], color='green', marker='o', s=30, alpha=0.6, label="Bronze")
 desc_stats = data_EU['total medals'].describe()
 print("Descriptive Statistics for total medals:")
 print(desc_stats)
 ax.tick_params(axis='both', labelsize=14)
 ax.grid(False)
 ax.set_xlabel('Medal Categories', fontsize=14)
 ax.set_ylabel('Total Medals', fontsize=14)
 ax.set_title(region+" in 2023 Olympics", fontsize=12)
 plt.legend(title="Medal Categories", bbox_to_anchor=(1, 0.8), loc='center right')
 plt.show()

plot_medals(data, "Europe")
```

```python
#bar chart
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("/Users/bharathi/Documents/sundar/olympics.csv")
data.columns = data.columns.str.strip()

def bar_region(data):
 region_medal_totals = data.groupby("region")[["gold", "silver", "bronze"]].sum()
 region_medal_totals.plot(kind="bar", stacked=True, color=["gold", "silver", "brown"], figsize=(6, 6))
 median_Tmedals = data['total medals'].median()
 print(f"Median for Total Medals: {median_Tmedals}")   
 std_gdp = data['total medals'].std()
 print(f"Standard Deviation of Total Medals {std_gdp}")
 plt.xlabel("Region")
 plt.ylabel("Total Medals")
 plt.title("Total Count of Each Medal Type by Region")
 plt.legend(title="Medal Categories")
 plt.show()
bar_region(data)  
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("/Users/bharathi/Documents/sundar/olympics.csv")
data.columns = data.columns.str.strip()
def box_gdp(data):

 plt.figure(figsize=(6, 5))
 sns.boxplot(x='region', y='gdp', data=data)
 plt.xticks(rotation=90) 
 plt.title("Box Plot of GDP by Region", fontsize=16)
 plt.xlabel("Region", fontsize=14)
 plt.ylabel("GDP", fontsize=14)
 skew_gdp = data['gdp'].skew()
 kurt_gdp = data['gdp'].kurtosis()
 print(f"Skewness of GDP: {skew_gdp}")
 print(f"Kurtosis of GDP: {kurt_gdp}")   
 gdp_population_corr = data[['gdp', 'population']].corr()
 print(f"Corelation:{gdp_population_corr}")   
 plt.show()
    
box_gdp(data)
```
